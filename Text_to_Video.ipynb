```python
import os
from gtts import gTTS
from moviepy.editor import ImageClip, AudioFileClip, concatenate_audioclips, concatenate_videoclips

# 1Ô∏è‚É£ Story Text Input
story = [
    "Once upon a time, in a faraway kingdom, there lived a brave knight.",
    "The knight embarked on a quest to find the lost treasure hidden deep in the enchanted forest.",
    "After facing many challenges, the knight finally discovered the treasure and returned as a hero."
]

# 2Ô∏è‚É£ Convert Story to Speech
audio_clips = []
image_paths = ["/content/TEMP/real_00001.jpg", "/content/TEMP/real_00002.jpg", "/content/TEMP/real_00003.jpg"]  # Replace with actual images
output_audio_paths = []

for i, paragraph in enumerate(story):
    tts = gTTS(paragraph)
    audio_path = f"audio_{i}.mp3"
    tts.save(audio_path)
    output_audio_paths.append(audio_path)
    audio_clips.append(AudioFileClip(audio_path))

# 3Ô∏è‚É£ Combine Audio Files (Fixed)
full_audio = concatenate_audioclips(audio_clips)

# 4Ô∏è‚É£ Create Video Clips with Images
video_clips = []
for i, img_path in enumerate(image_paths):
    img_clip = ImageClip(img_path, duration=audio_clips[i].duration)
    img_clip = img_clip.set_audio(audio_clips[i])  # Attach audio to the image
    video_clips.append(img_clip)

# 5Ô∏è‚É£ Merge Everything into a Final Video
final_video = concatenate_videoclips(video_clips)
final_video = final_video.set_audio(full_audio)  # Ensure audio is attached

output_video_path = "/content/TEMP/story_video.mp4"
final_video.write_videofile(output_video_path, fps=24)

print("üé• Story Video Created Successfully!")
```

```python
import os
import random
from gtts import gTTS
from moviepy.editor import (
    ImageClip, AudioFileClip, concatenate_audioclips, concatenate_videoclips,
    TextClip, CompositeVideoClip, CompositeAudioClip
)
from diffusers import StableDiffusionPipeline
import torch

# Load Stable Diffusion model (Run this once)
pipe = StableDiffusionPipeline.from_pretrained("runwayml/stable-diffusion-v1-5").to("cuda" if torch.cuda.is_available() else "cpu")

# 1Ô∏è‚É£ Story Text Input
story = [
    "Once upon a time, in a faraway kingdom, there lived a brave knight.",
    "The knight embarked on a quest to find the lost treasure hidden deep in the enchanted forest.",
    "After facing many challenges, the knight finally discovered the treasure and returned as a hero."
]

# 2Ô∏è‚É£ Generate AI Images using Stable Diffusion
image_paths = []
for i, paragraph in enumerate(story):
    prompt = f"Fantasy illustration of {paragraph}, cinematic, highly detailed"
    image = pipe(prompt).images[0]  # Generate image
    img_path = f"scene_{i}.png"
    image.save(img_path)  # Save locally
    image_paths.append(img_path)

# 3Ô∏è‚É£ Convert Story to Speech
audio_clips = []
output_audio_paths = []

for i, paragraph in enumerate(story):
    tts = gTTS(paragraph)
    audio_path = f"audio_{i}.mp3"
    tts.save(audio_path)
    output_audio_paths.append(audio_path)
    audio_clips.append(AudioFileClip(audio_path))

# 4Ô∏è‚É£ Combine Audio Files
full_audio = concatenate_audioclips(audio_clips)



# 6Ô∏è‚É£ Create Video Clips with AI Images and Text Animations

```

```python
# 5Ô∏è‚É£ Add Background Music
bg_music = AudioFileClip("/content/TEMP/marimba-ambient-music-301293.mp3").set_duration(full_audio.duration).volumex(0.3)
final_audio = CompositeAudioClip([full_audio, bg_music])
video_clips = []
for i, img_path in enumerate(image_paths):
    img_clip = ImageClip(img_path, duration=audio_clips[i].duration)
    img_clip = img_clip.set_audio(audio_clips[i])  # Attach audio to the image

    # Add text animation
    text_clip = TextClip(story[i], fontsize=50, color='white', size=img_clip.size, method='caption')
    text_clip = text_clip.set_position('center').set_duration(audio_clips[i].duration)

    # Merge text and image
    final_scene = CompositeVideoClip([img_clip, text_clip])
    video_clips.append(final_scene)

# 7Ô∏è‚É£ Merge Everything into a Final Video
final_video = concatenate_videoclips(video_clips)
final_video = final_video.set_audio(final_audio)  # Attach final audio

output_video_path = "/content/TEMP/story_video.mp4"
final_video.write_videofile(output_video_path, fps=24)

print("üé• AI-Generated Story Video Created Successfully!")
```
